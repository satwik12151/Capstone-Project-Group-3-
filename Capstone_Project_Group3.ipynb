# PHASE 1 — Revenue Base & Trend Decomposition

Revenue trend analysis is the foundation of any business diagnosis — it tells you not just how much you're selling, but whether the business is genuinely growing or riding seasonal waves. Phase 1 separates real structural growth from calendar noise, revealing which customer segments are driving momentum and whether that momentum is building or quietly fading.

## Question 1 — Monthly Revenue Trend

Is revenue consistently climbing year-on-year, or are seasonal spikes masking an underlying slowdown?

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import matplotlib.gridspec as gridspec
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")

# ── Colour palette
C = {
    "blue":   "#378ADD",
    "coral":  "#D85A30",
    "teal":   "#1D9E75",
    "purple": "#7F77DD",
    "amber":  "#BA7517",
    "gray":   "#888780",
}

plt.rcParams.update({
    "figure.facecolor": "white",
    "axes.facecolor":   "white",
    "axes.grid":        True,
    "grid.alpha":       0.3,
    "axes.spines.top":  False,
    "axes.spines.right":False,
    "font.size":        11,
})

# ── Prepare monthly revenue
df["order_date"] = pd.to_datetime(df["Order Date"])
df["order_ym"]   = df["order_date"].dt.to_period("M")

monthly_rev = (
    df.groupby("order_ym")["Sales"]
    .sum()
    .reset_index()
    .rename(columns={"Sales": "monthly_revenue"})
)
monthly_rev["rolling_90d"] = monthly_rev["monthly_revenue"].rolling(3).mean()
monthly_rev["ds"] = monthly_rev["order_ym"].dt.to_timestamp()

# ── Plot
fig, ax = plt.subplots(figsize=(16, 5))

ax.plot(monthly_rev["ds"], monthly_rev["monthly_revenue"],
        color=C["blue"], lw=2, label="Monthly revenue")
ax.plot(monthly_rev["ds"], monthly_rev["rolling_90d"],
        color=C["coral"], lw=1.5, ls="--", label="3-month rolling avg")
ax.fill_between(monthly_rev["ds"], monthly_rev["monthly_revenue"],
                alpha=0.1, color=C["blue"])
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"${x:,.0f}"))
ax.set_title("Monthly Revenue Trend — Is growth real or seasonal?", fontweight="bold")
ax.legend(fontsize=9)

# ── Annotate peak and trough
peak_idx   = monthly_rev["monthly_revenue"].idxmax()
trough_idx = monthly_rev["monthly_revenue"].idxmin()
peak_row   = monthly_rev.loc[peak_idx]
trough_row = monthly_rev.loc[trough_idx]

ax.annotate(f"Peak: ${peak_row['monthly_revenue']:,.0f}",
            xy=(peak_row["ds"], peak_row["monthly_revenue"]),
            xytext=(10, -20), textcoords="offset points",
            fontsize=8, color=C["teal"],
            arrowprops=dict(arrowstyle="->", color=C["teal"], lw=1))
ax.annotate(f"Low: ${trough_row['monthly_revenue']:,.0f}",
            xy=(trough_row["ds"], trough_row["monthly_revenue"]),
            xytext=(10, 15), textcoords="offset points",
            fontsize=8, color=C["coral"],
            arrowprops=dict(arrowstyle="->", color=C["coral"], lw=1))

plt.tight_layout()
plt.show()

# ── Key output
print(f"Total revenue : ${df['Sales'].sum():,.0f}")
print(f"Peak month    : {peak_row['ds'].strftime('%b %Y')} — ${peak_row['monthly_revenue']:,.0f}")
print(f"Lowest month  : {trough_row['ds'].strftime('%b %Y')} — ${trough_row['monthly_revenue']:,.0f}")
```

## Inference — Monthly Revenue Trend

Revenue grew from **$91,152 in February 2011** to a peak of **$555,279 in November 2014** — a 6x increase over four years, confirming genuine structural growth.

The rolling average tracks closely with actual revenue, meaning growth is consistent and not driven by one-off spikes. A seasonal dip in Q1 and a surge in Q4 is visible every year, but even the weakest months in later years outperform the best months in earlier years.

## Question 2 — Revenue by Segment over Time

Which customer segment — Consumer, Corporate, or Home Office — is carrying the business? And is that changing over time?

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import warnings
warnings.filterwarnings("ignore")

# ── Colour palette
C = {
    "blue":   "#378ADD",
    "coral":  "#D85A30",
    "teal":   "#1D9E75",
    "purple": "#7F77DD",
    "amber":  "#BA7517",
    "gray":   "#888780",
}
SEGMENT_PALETTE = {
    "Consumer":    C["blue"],
    "Corporate":   C["teal"],
    "Home Office": C["amber"],
}

# ── Prepare data
df["order_date"] = pd.to_datetime(df["Order Date"])
df["order_ym"]   = df["order_date"].dt.to_period("M")

seg_rev = df.groupby(["order_ym", "Segment"])["Sales"].sum().reset_index()
seg_pivot = seg_rev.copy()
seg_pivot["ds"] = seg_pivot["order_ym"].dt.to_timestamp()
seg_pivot = seg_pivot.pivot(index="ds", columns="Segment", values="Sales").fillna(0)

# ── Plot
fig, ax = plt.subplots(figsize=(16, 6))
seg_pivot.plot.area(
    ax=ax,
    color=[SEGMENT_PALETTE.get(c, C["gray"]) for c in seg_pivot.columns],
    alpha=0.75,
    linewidth=0
)
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"${x:,.0f}"))
ax.set_title("Revenue by Segment — Which segment is carrying the business?", fontweight="bold")
ax.legend(fontsize=9, loc="upper left")
ax.set_xlabel("")

# ── % share text box
seg_share = df.groupby("Segment")["Sales"].sum()
seg_share_pct = (seg_share / seg_share.sum() * 100).round(1)
share_text = "\n".join([f"{seg}: {pct}%" for seg, pct in seg_share_pct.items()])
ax.text(0.99, 0.95, share_text,
        transform=ax.transAxes, fontsize=9,
        verticalalignment="top", horizontalalignment="right",
        bbox=dict(boxstyle="round,pad=0.4", facecolor="white", alpha=0.8))

plt.tight_layout()
plt.show()

# ── Key output
top_seg = seg_share_pct.idxmax()
print(f"Segment revenue share:")
for seg, pct in seg_share_pct.items():
    print(f"  {seg}: {pct}%")
print(f"\nBusiness insight: {top_seg} dominates with {seg_share_pct.max()}% of total revenue.")
```

## Inference — Revenue by Segment

Consumer is the dominant segment, contributing the largest share of total revenue, followed by Corporate and Home Office.

All three segments show growth over time, but Consumer consistently leads. The stacked area chart shows no major shift in segment mix — the business is not becoming over-reliant on one segment, which reduces concentration risk.

**Bottom line:** Consumer is the engine of growth. Corporate is a stable second. Home Office is the smallest but steady — worth monitoring for upsell opportunities.

## Question 3 — YoY Revenue Growth Rate

Is growth accelerating or decelerating? Are we consistently beating last year, or are gains concentrated in a few lucky months?

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import warnings
warnings.filterwarnings("ignore")

# ── Colour palette
C = {
    "teal":   "#1D9E75",
    "coral":  "#D85A30",
    "purple": "#7F77DD",
    "gray":   "#888780",
}

# ── Prepare monthly revenue
df["order_date"] = pd.to_datetime(df["Order Date"])
df["order_ym"]   = df["order_date"].dt.to_period("M")

monthly_rev = (
    df.groupby("order_ym")["Sales"]
    .sum()
    .reset_index()
    .rename(columns={"Sales": "monthly_revenue"})
)

# ── Calculate YoY growth
monthly_rev["order_year"]  = monthly_rev["order_ym"].dt.year
monthly_rev["order_month"] = monthly_rev["order_ym"].dt.month

prev = monthly_rev.copy()
prev["order_year"] += 1
monthly_rev = monthly_rev.merge(
    prev[["order_year", "order_month", "monthly_revenue"]],
    on=["order_year", "order_month"], how="left", suffixes=("", "_prev")
)
monthly_rev["yoy_growth_rate"] = (
    (monthly_rev["monthly_revenue"] - monthly_rev["monthly_revenue_prev"])
    / monthly_rev["monthly_revenue_prev"].replace(0, None)
) * 100

yoy_clean = monthly_rev.dropna(subset=["yoy_growth_rate"]).copy()
yoy_clean["ds"] = yoy_clean["order_ym"].dt.to_timestamp()

# ── Plot
fig, ax = plt.subplots(figsize=(16, 5))

colours_yoy = [C["teal"] if v >= 0 else C["coral"] for v in yoy_clean["yoy_growth_rate"]]
ax.bar(yoy_clean["ds"], yoy_clean["yoy_growth_rate"], color=colours_yoy, width=20)
ax.axhline(0, color=C["gray"], lw=0.8)

# ── Average growth line
avg_growth = yoy_clean["yoy_growth_rate"].mean()
ax.axhline(avg_growth, color=C["purple"], lw=1.5, ls="--",
           label=f"Avg growth: {avg_growth:.1f}%")

ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"{x:.0f}%"))
ax.set_title("YoY Revenue Growth Rate — Is growth accelerating or decelerating?", fontweight="bold")
ax.legend(fontsize=9)

plt.tight_layout()
plt.show()

# ── Key output
positive_months = (yoy_clean["yoy_growth_rate"] > 0).sum()
total_months    = len(yoy_clean)
print(f"Avg YoY growth     : {avg_growth:.1f}%")
print(f"Positive months    : {positive_months}/{total_months}")
print(f"Business insight   : {positive_months}/{total_months} months beat last year — "
      f"{'healthy momentum' if positive_months > total_months/2 else 'growth is inconsistent'}.")
```

## Inference — YoY Revenue Growth Rate

35 out of 36 months beat the previous year, with an average YoY growth rate of **26.4%**. Growth is not concentrated in a few lucky months — it is broad-based and consistent across the entire period.

The bars are predominantly green, with only one negative month, confirming that the business has maintained momentum year after year. The average growth line at 26.4% sits comfortably above zero, reinforcing the structural upward trend seen in Question 1.

**Bottom line:** Growth is real, consistent, and strong. A 26.4% average YoY growth rate over 3 years is exceptional for a retail business and sets a high but credible benchmark for future targets.

## **PHASE 2 — Average Order Value Analysis**

Average Order Value (AOV) is the clearest proxy for revenue quality — it tells you not just how much you're selling but how well each transaction is performing. Phase 2 separates high-value, healthy orders from discounted, inflated ones, revealing where margin is being created or quietly destroyed across segments and time.

```python
import warnings, numpy as np, pandas as pd
import matplotlib.pyplot as plt, matplotlib.ticker as mticker
warnings.filterwarnings("ignore")

SEG_PALETTE = {"Consumer": "#378ADD", "Corporate": "#1D9E75", "Home Office": "#BA7517"}
plt.rcParams.update({
    "figure.facecolor": "white", "axes.facecolor": "white",
    "axes.grid": True, "grid.alpha": 0.3,
    "axes.spines.top": False, "axes.spines.right": False, "font.size": 11,
})
```

```python
orders = df.groupby("Order ID").agg(
    aov=("Sales", "sum"),
    order_profit=("Profit", "sum"),
    avg_discount=("Discount", "mean"),
    order_date=("Order Date", "first"),
    segment=("Segment", "first"),
    category=("Category", lambda x: x.value_counts().index[0]),
).reset_index()

# Convert dates
orders["order_date"] = pd.to_datetime(orders["order_date"])

# Create time columns
orders["order_ym"] = orders["order_date"].dt.to_period("M")
orders["ds"] = orders["order_ym"].dt.to_timestamp()
orders["order_year"] = orders["ds"].dt.year

# Profit margin
orders["profit_margin"] = (
    orders["order_profit"] /
    orders["aov"].replace(0, np.nan)
)

# Discount bands
orders["discount_band"] = pd.cut(
    orders["avg_discount"],
    bins=[-0.001, 0, 0.1, 0.2, 0.3, 1.0],
    labels=["0%", "1-10%", "11-20%", "21-30%", "31%+"]
)

# High AOV flag
threshold = orders["aov"].median()

orders["high_aov_flag"] = (
    orders["aov"] >= threshold
).astype(int)

print(
    f"Orders built: {len(orders):,} | "
    f"High-AOV threshold (median): ${threshold:,.2f}"
)
```

## Question 1

**Which segment places the highest-value orders — and is that gap widening or closing?**

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle(
    "Q1 — Which segment places the highest-value orders — and is that gap widening or closing?",
    fontsize=12, fontweight="bold"
)

seg_order = orders.groupby("segment")["aov"].median().sort_values(ascending=False).index.tolist()
parts = axes[0].violinplot(
    [orders[orders["segment"] == s]["aov"].values for s in seg_order],
    positions=range(len(seg_order)), showmedians=True, showextrema=False
)
for pc, seg in zip(parts["bodies"], seg_order):
    pc.set_facecolor(SEG_PALETTE.get(seg, "#888")); pc.set_alpha(0.6)
parts["cmedians"].set_color("#222")
axes[0].set_xticks(range(len(seg_order))); axes[0].set_xticklabels(seg_order)
axes[0].yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"${x:,.0f}"))
axes[0].set_ylabel("Order value ($)")
axes[0].set_title("AOV distribution by segment", fontweight="bold")
for i, seg in enumerate(seg_order):
    med = orders[orders["segment"] == seg]["aov"].median()
    axes[0].text(i, med + 20, f"${med:,.0f}", ha="center", fontsize=9, fontweight="bold")

for seg, colour in SEG_PALETTE.items():
    ts = (orders[orders["segment"] == seg]
          .groupby("ds")["aov"].median().sort_index().rolling(3, min_periods=1).mean())
    axes[1].plot(ts.index, ts.values, color=colour, lw=2, label=seg)
axes[1].yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"${x:,.0f}"))
axes[1].set_title("Median AOV (3-month rolling) by segment", fontweight="bold")
axes[1].legend(fontsize=9)

fig.tight_layout()
plt.show()

print("\nSegment AOV summary:")
print(orders.groupby("segment")["aov"].agg(
    Median="median", Mean="mean", P75=lambda x: x.quantile(0.75)
).round(2).to_string())
```

### Inference

Consumer places the highest-value orders, with the top median AOV at approximately **$211**, slightly ahead of Corporate (**$205**) and Home Office (**$201**).

The gap, however, is narrow and not structurally widening. The **3-month rolling median AOV trends** show all three segments moving within a similar range, with periodic crossovers indicating short-term volatility rather than sustained divergence.

Corporate also demonstrates the strongest upper-end order potential, with the highest **75th percentile (P75 ≈ $633)**, suggesting greater premium-order upside.

Overall, while segment-level AOV differences do exist, they remain modest. This suggests that **broad-based basket expansion strategies** may deliver more impact than highly segment-specific pricing approaches.

## Question 2

**At what discount level does AOV collapse — are we systematically over-discounting?**

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle(
    "Q2 — At what discount level does AOV collapse — are we systematically over-discounting?",
    fontsize=12, fontweight="bold"
)

band_stats = (
    orders.groupby("discount_band", observed=True)
    .agg(avg_aov=("aov","mean"), avg_margin=("profit_margin","mean"), n=("aov","count"))
    .reset_index()
)
x    = range(len(band_stats))
ax_r = axes[0].twinx()
bars = axes[0].bar(x, band_stats["avg_aov"],
                   color=["#378ADD" if i < 2 else "#D85A30" for i in x],
                   alpha=0.75, width=0.5)
ax_r.plot(x, band_stats["avg_margin"] * 100, color="#1D9E75", marker="o", lw=2)
ax_r.axhline(0, color="#888", lw=0.8, ls="--")
axes[0].set_xticks(list(x)); axes[0].set_xticklabels(band_stats["discount_band"].astype(str))
axes[0].yaxis.set_major_formatter(mticker.FuncFormatter(lambda v, _: f"${v:,.0f}"))
axes[0].set_ylabel("Avg AOV ($)", color="#378ADD")
ax_r.set_ylabel("Avg margin (%)", color="#1D9E75")
axes[0].set_title("AOV and margin by discount band", fontweight="bold")
for bar, row in zip(bars, band_stats.itertuples()):
    axes[0].text(bar.get_x() + bar.get_width()/2, bar.get_height() + 5,
                 f"n={row.n}", ha="center", fontsize=8, color="#555")

sample  = orders.sample(min(2000, len(orders)), random_state=42)
colours = ["#D85A30" if p < 0 else "#378ADD" for p in sample["order_profit"]]
axes[1].scatter(sample["avg_discount"] * 100, sample["aov"],
                c=colours, alpha=0.35, s=14)
axes[1].axvline(20, color="#D85A30", ls="--", lw=1.2, label="20% discount line")
axes[1].set_xlabel("Discount (%)"); axes[1].set_ylabel("Order value ($)")
axes[1].yaxis.set_major_formatter(mticker.FuncFormatter(lambda v, _: f"${v:,.0f}"))
axes[1].set_title("AOV vs discount — red = loss-making orders", fontweight="bold")
axes[1].legend(fontsize=9)

fig.tight_layout()
plt.show()

loss = (orders.groupby("discount_band", observed=True)
        .apply(lambda x: (x["order_profit"] < 0).mean() * 100).reset_index())
loss.columns = ["Discount band", "Loss-making (%)"]
print("\nLoss-making order rate by discount band:")
print(loss.round(1).to_string(index=False))
```

## Inference

- Discounting beyond **20%** appears to be **value-destructive rather than growth-accretive**.

- While AOV remains elevated in the **21–30% discount band** (~$730), profitability deteriorates sharply:
  - Average margin falls below zero
  - Approximately **52% of orders become loss-making**

- The impact worsens significantly in the **31%+ discount band**:
  - Average margin declines further to approximately **-59%**
  - Nearly **91% of orders are unprofitable**

- The scatter plot further reinforces this relationship, with loss-making orders clustering heavily above the **20% discount threshold**.

- Overall, the analysis suggests the business may be **systematically over-discounting beyond 20%**.

> **Tighter discount guardrails and stronger discount governance appear necessary to protect profitability.**

## Question 3

**What share of orders clear the high-quality AOV threshold — and which categories drive them?**

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle(
    f"Q3 — What share of orders clear the high-quality AOV threshold (>=${threshold:,.0f})"
    f" — and which categories drive them?",
    fontsize=12, fontweight="bold"
)

seg_counts = orders.groupby(["segment", "high_aov_flag"]).size().unstack(fill_value=0)
seg_pct    = seg_counts.div(seg_counts.sum(axis=1), axis=0) * 100
for col in [0, 1]:
    if col not in seg_pct.columns: seg_pct[col] = 0
seg_pct[[1, 0]].rename(
    columns={1: f">=${threshold:,.0f} (high)", 0: f"<${threshold:,.0f} (low)"}
).plot.barh(ax=axes[0], stacked=True, color=["#1D9E75", "#D3D1C7"], edgecolor="none", width=0.5)
axes[0].set_xlabel("% of orders")
axes[0].set_title("High-AOV share by segment", fontweight="bold")
axes[0].legend(fontsize=8, loc="lower right")
for i, seg in enumerate(seg_pct.index):
    pct = seg_pct.loc[seg, 1] if 1 in seg_pct.columns else 0
    axes[0].text(pct / 2, i, f"{pct:.0f}%", va="center", ha="center",
                 fontsize=9, color="white", fontweight="bold")

cat_aov = orders.groupby(["category", "high_aov_flag"])["aov"].mean().unstack(fill_value=0)
for col in [0, 1]:
    if col not in cat_aov.columns: cat_aov[col] = 0
cat_aov.sort_values(1, ascending=True).rename(
    columns={0: f"<${threshold:,.0f}", 1: f">=${threshold:,.0f}"}
).plot.barh(ax=axes[1], color=["#D3D1C7", "#7F77DD"], edgecolor="none", width=0.5)
axes[1].xaxis.set_major_formatter(mticker.FuncFormatter(lambda v, _: f"${v:,.0f}"))
axes[1].set_title("Avg AOV by category — high vs low quality orders", fontweight="bold")
axes[1].legend(fontsize=8)

fig.tight_layout()
plt.show()

print("\nHigh-AOV share by segment:")
print(orders.groupby("segment")["high_aov_flag"].mean()
      .mul(100).round(1).rename("% high-AOV orders").to_string())
```

### Inference

High-quality orders (AOV $\geq$ \$207) account for approximately half of all volume, showing near-identical penetration across Consumer (50.4%), Corporate (49.9%), and Home Office (49.1%) segments. This suggests segment-agnostic buying behavior at the \$207 threshold. Among these high-value orders, Technology leads strongly, pushing average AOV above \$1,000, followed closely by Furniture and Office Supplies. To elevate the overall order mix, efforts should focus on cross-selling Technology products within low-value carts rather than segment-specific targeting.

## Question 4

**Is per-segment AOV trending up or down — an early warning of pricing or loyalty erosion?**

```python
fig, axes = plt.subplots(2, 1, figsize=(14, 9))
fig.suptitle(
    "Q4 — Is per-segment AOV trending up or down? Early warning of pricing or loyalty erosion.",
    fontsize=12, fontweight="bold"
)

for seg, colour in SEG_PALETTE.items():
    ts = orders[orders["segment"] == seg].groupby("ds")["aov"].mean().sort_index()
    axes[0].plot(ts.index, ts.values, color=colour, lw=1.2, alpha=0.5)
    if len(ts) > 2:
        coef = np.polyfit(np.arange(len(ts)), ts.values, 1)
        axes[0].plot(ts.index, np.poly1d(coef)(np.arange(len(ts))),
                     color=colour, lw=2.5, label=f"{seg}  ({coef[0]:+.1f} $/mo)")
axes[0].yaxis.set_major_formatter(mticker.FuncFormatter(lambda v, _: f"${v:,.0f}"))
axes[0].set_title("Monthly mean AOV with linear trend per segment", fontweight="bold")
axes[0].legend(fontsize=9)

yr_seg = orders.groupby(["order_year","segment"])["aov"].mean().unstack()
yoy    = yr_seg.pct_change().mul(100).dropna(how="all")
x      = np.arange(len(yoy))
for i, (seg, colour) in enumerate(SEG_PALETTE.items()):
    if seg not in yoy.columns: continue
    vals = yoy[seg].fillna(0)
    axes[1].bar(x + i * 0.22, vals,
                width=0.22, edgecolor="none", label=seg,
                color=[colour if v >= 0 else "#D85A30" for v in vals], alpha=0.85)
axes[1].axhline(0, color="#888", lw=0.8)
axes[1].set_xticks(x + 0.22); axes[1].set_xticklabels([str(y) for y in yoy.index])
axes[1].yaxis.set_major_formatter(mticker.FuncFormatter(lambda v, _: f"{v:+.0f}%"))
axes[1].set_title("YoY change in mean AOV by segment", fontweight="bold")
axes[1].set_xlabel("Year"); axes[1].legend(fontsize=9)

fig.tight_layout()
plt.show()

print("\nTrend slope ($/month) — negative = eroding AOV:")
for seg in SEG_PALETTE:
    ts = orders[orders["segment"] == seg].groupby("ds")["aov"].mean().sort_index()
    if len(ts) > 2:
        slope = np.polyfit(np.arange(len(ts)), ts.values, 1)[0]
        print(f"  {seg:15s}: {slope:+.2f} $/month  [{'DECLINING' if slope < 0 else 'growing'}]")
```

### Inference

Per-segment AOV shows no systemic erosion, remaining largely flat over the multi-year period with only marginal long-term upward slopes (+$0.03 to +$0.49/month). However, year-over-year performance is highly volatile. Corporate AOV faced sharp contractions in 2012 and 2014, while Home Office saw a notable ~8% spike in 2014 after previous declines. Despite dramatic monthly fluctuations—including early-2011 outlier spikes for Corporate and Home Office—the baseline AOV holds steady at ~$500 across all segments. Strategic focus should shift from defending baseline pricing to dampening year-to-year volatility and driving structural basket growth.

## **PHASE 3 — CUSTOMER SEGMENTATION & PROFILING**
Revenue numbers look healthy on the surface — but revenue alone doesn't tell us *who* is actually driving the business. The Global Superstore serves three named customer segments: **Consumer**, **Corporate**, and **Home Office**.

This phase asks the critical question: **are these segments truly distinct in their buying behaviour, profitability, and risk — or are they just labels?**

### Question 2
**Is the loss rate uniform across segments — or is one segment significantly more likely to generate unprofitable orders?**

```python
# Compute loss stats per segment
loss_summary = df.groupby('Segment').apply(
    lambda x: pd.Series({
        'Total Orders': len(x),
        'Loss Orders': (x['Profit'] < 0).sum(),
        'Loss Rate (%)': round((x['Profit'] < 0).mean() * 100, 1),
        'Avg Profit (USD)': round(x['Profit'].mean(), 2),
        'Median Profit (USD)': round(x['Profit'].median(), 2)
    })
).reset_index()

print("--- Loss Rate by Segment ---")
print(loss_summary.to_string(index=False))

# Stacked bar — profitable vs loss orders
fig, ax = plt.subplots(figsize=(10, 5))

segments = loss_summary['Segment']
loss_orders = loss_summary['Loss Orders']
profit_orders = loss_summary['Total Orders'] - loss_orders

x = range(len(segments))
bars_p = ax.bar(x, profit_orders, label='Profitable Orders', color='#2ecc71', edgecolor='white')
bars_l = ax.bar(x, loss_orders, bottom=profit_orders, label='Loss Orders', color='#e74c3c', edgecolor='white')

ax.set_xticks(x)
ax.set_xticklabels(segments, fontsize=12)
ax.set_title('Profitable vs Loss-Making Orders by Segment', fontsize=14, fontweight='bold', pad=12)
ax.set_ylabel('Number of Orders', fontsize=11)
ax.legend(fontsize=11)

# Annotate loss rate on top
for i, (bar, rate) in enumerate(zip(bars_l, loss_summary['Loss Rate (%)'])):
    ax.text(bar.get_x() + bar.get_width()/2,
            profit_orders.iloc[i] + loss_orders.iloc[i] + 300,
            f'Loss Rate: {rate}%', ha='center', fontsize=10,
            fontweight='bold', color='#c0392b')

plt.tight_layout()
plt.show()
```

###  Inference — Question 2

- Consumer: **24.6%**
- Corporate: **24.4%**
- Home Office: **24.1%**

Roughly **1 in 4 orders loses money in every segment** without exception.

This uniformity is a significant finding. It tells us the loss problem is **not a segment-level behaviour** . Solving it requires looking at product pricing or discount policy, not at which type of customer is buying.

###  Question 3
**Who are our top individual customers by total profit — and does a Pareto pattern exist (do 20% of customers drive 80% of profit)?**

```python
# Customer-level profit aggregation
customer_profit = (df.groupby(['Customer ID', 'Customer Name', 'Segment'])['Profit']
                   .sum()
                   .reset_index()
                   .sort_values('Profit', ascending=False))

# Top 15 customers
top15 = customer_profit.head(15)

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# --- Plot 1: Top 15 customers horizontal bar ---
seg_colors = {'Consumer': '#4C72B0', 'Corporate': '#DD8452', 'Home Office': '#55A868'}
bar_colors = [seg_colors[s] for s in top15['Segment']]

axes[0].barh(top15['Customer Name'][::-1], top15['Profit'][::-1],
             color=bar_colors[::-1], edgecolor='white')
axes[0].set_title('Top 15 Customers by Total Profit', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Total Profit (USD)', fontsize=11)

# Add legend
from matplotlib.patches import Patch
legend_elements = [Patch(facecolor=v, label=k) for k, v in seg_colors.items()]
axes[0].legend(handles=legend_elements, fontsize=10)

# --- Plot 2: Pareto Curve ---
all_customers_sorted = customer_profit.sort_values('Profit', ascending=False)
cumulative_profit = all_customers_sorted['Profit'].cumsum()
total_profit = all_customers_sorted['Profit'].sum()
cumulative_pct = cumulative_profit / total_profit * 100
customer_pct = np.array(range(1, len(all_customers_sorted) + 1)) / len(all_customers_sorted) * 100

axes[1].plot(customer_pct, cumulative_pct, color='#2980b9', linewidth=2)
axes[1].axhline(80, color='red', linestyle='--', linewidth=1, label='80% profit line')
axes[1].axvline(20, color='orange', linestyle='--', linewidth=1, label='20% customers line')
axes[1].fill_between(customer_pct, cumulative_pct, alpha=0.1, color='#2980b9')
axes[1].set_title('Pareto Curve — Customer Profit Concentration', fontsize=13, fontweight='bold')
axes[1].set_xlabel('Cumulative % of Customers', fontsize=11)
axes[1].set_ylabel('Cumulative % of Total Profit', fontsize=11)
axes[1].legend(fontsize=10)

# Find actual 80/20 point
idx_80 = (cumulative_pct >= 80).idxmax()
cust_pct_at_80 = customer_pct[all_customers_sorted.index.get_loc(idx_80)]
print(f"80% of profit is generated by top {cust_pct_at_80:.1f}% of customers")

plt.tight_layout()
plt.show()

print("\n--- Top 10 Customers by Profit ---")
print(top15[['Customer Name', 'Segment', 'Profit']].head(10).to_string(index=False))
```

### Inference — Question 3

The top 15 customers are drawn from all three segments —
confirming that high-value customers are not concentrated
in any one segment, they exist across the entire base.

**The Pareto curve shows a 29/80 pattern** — 80% of total
profit is driven by the top 29.4% of customers. This is
weaker than the classic 80/20 rule, meaning profit is
somewhat more evenly spread than typical retail businesses.
However, nearly 70% of customers still contribute only
the remaining 20% of profit — a long tail of low-value accounts.

**The curve dropping beyond its peak** reveals that a portion
of customers are loss-making — their orders are so deeply
unprofitable that they actively reduce cumulative profit.
These customers cost the business money to serve.

**Strategic implication:** The top customer tier deserves
retention focus, but the loss-making tail at the other end
is equally important — identifying and addressing those
accounts could recover margin without acquiring a single
new customer.

# **PHASE 5 - Revenue Forecasting by Segment × Market**

Revenue forecasting is where descriptive analysis becomes a decision-making instrument. The first four phases tell you what happened; Phase 5 tells you what to prepare for. Without it, budgets are guesses, hiring decisions are reactive, and inventory planning is based on hope. Breaking the forecast by segment reveals which part of the business is actually growing versus which is masking decline in the aggregate. The STL decomposition prevents the most common planning mistake in retail — mistaking seasonal lift for structural growth. The leading indicators analysis gives teams an early warning system, turning a monthly surprise into a weekly signal they can act on.

```python
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
from statsmodels.tsa.seasonal import STL
from statsmodels.tsa.statespace.sarimax import SARIMAX

warnings.filterwarnings("ignore")

SEG_PALETTE = {"Consumer": "#378ADD", "Corporate": "#1D9E75", "Home Office": "#BA7517"}
FORECAST_PERIODS = 6   # months ahead — adjust as needed

plt.rcParams.update({
    "figure.facecolor": "white", "axes.facecolor": "white",
    "axes.grid": True, "grid.alpha": 0.3,
    "axes.spines.top": False, "axes.spines.right": False, "font.size": 11,
})
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["order_ym"]   = df["Order Date"].dt.to_period("M")

# overall monthly revenue
ts = (
    df.groupby("order_ym")["Sales"].sum()
    .sort_index()
    .asfreq("M")
    .fillna(method="ffill")
)

# per-segment monthly revenue
seg_ts = {}
for seg in df["Segment"].unique():
    seg_ts[seg] = (
        df[df["Segment"] == seg]
        .groupby("order_ym")["Sales"].sum()
        .sort_index()
        .asfreq("M")
        .fillna(method="ffill")
    )

# lag-feature table for Q4
lag_df = pd.DataFrame({"revenue": ts})
lag_df["lag_1"]          = lag_df["revenue"].shift(1)
lag_df["lag_2"]          = lag_df["revenue"].shift(2)
lag_df["lag_3"]          = lag_df["revenue"].shift(3)
lag_df["rolling_3m_avg"] = lag_df["revenue"].shift(1).rolling(3).mean()
lag_df["month"]          = lag_df.index.month
lag_df["is_q4"]          = lag_df["month"].isin([10, 11, 12]).astype(int)
lag_df["mom_growth"]     = lag_df["revenue"].pct_change(1) * 100   # month-on-month %
lag_df["qoq_growth"]     = lag_df["revenue"].pct_change(3) * 100   # quarter-on-quarter %

print(f"Time series: {ts.index[0]}  to  {ts.index[-1]}  ({len(ts)} months)")
print(f"Segments found: {list(seg_ts.keys())}")
```

## QUESTION 1

What will total revenue look like next quarter — and what's the realistic floor and ceiling?

```python
try:
    # Ensure numeric series
    ts = pd.to_numeric(ts, errors="coerce").dropna()

    # Fit SARIMA
    model = SARIMAX(
        ts,
        order=(1, 1, 1),
        seasonal_order=(1, 1, 0, 12),
        enforce_stationarity=False,
        enforce_invertibility=False
    )

    result = model.fit(disp=False)

    # Forecast
    fc = result.get_forecast(steps=FORECAST_PERIODS)

    fc_mean = fc.predicted_mean.astype(float)

    fc_ci = fc.conf_int(alpha=0.2).astype(float)       # 80% CI
    fc_ci_wide = fc.conf_int(alpha=0.05).astype(float) # 95% CI

    # Convert timestamps
    hist_x = ts.index.to_timestamp()
    fc_x = fc_mean.index.to_timestamp()

    # =====================================================
    # PLOTS
    # =====================================================
    fig, axes = plt.subplots(2, 1, figsize=(14, 9))

    fig.suptitle(
        "Q1 — What will total revenue be next quarter — and what's the realistic floor and ceiling?",
        fontsize=12,
        fontweight="bold"
    )

    # =====================================================
    # TOP PLOT — Forecast with CI
    # =====================================================
    ax = axes[0]

    ax.fill_between(
        fc_x,
        fc_ci_wide.iloc[:, 0].values,
        fc_ci_wide.iloc[:, 1].values,
        alpha=0.10,
        color="#378ADD",
        label="95% CI"
    )

    ax.fill_between(
        fc_x,
        fc_ci.iloc[:, 0].values,
        fc_ci.iloc[:, 1].values,
        alpha=0.22,
        color="#378ADD",
        label="80% CI"
    )

    ax.plot(
        hist_x,
        ts.values,
        color="#444",
        lw=1.8,
        label="Actual"
    )

    ax.plot(
        fc_x,
        fc_mean.values,
        color="#378ADD",
        lw=2.2,
        ls="--",
        label="Forecast"
    )

    ax.axvline(
        hist_x[-1],
        color="#D85A30",
        lw=1,
        ls=":",
        label="Forecast start"
    )

    ax.yaxis.set_major_formatter(
        mticker.FuncFormatter(lambda v, _: f"${v:,.0f}")
    )

    ax.set_title(
        "SARIMA Revenue Forecast with Confidence Intervals",
        fontweight="bold"
    )

    ax.legend(fontsize=9)

    # =====================================================
    # BOTTOM PLOT — Scenario ranges
    # =====================================================
    ax2 = axes[1]

    x = np.arange(FORECAST_PERIODS)

    ax2.bar(
        x,
        fc_ci_wide.iloc[:, 1].values - fc_ci_wide.iloc[:, 0].values,
        bottom=fc_ci_wide.iloc[:, 0].values,
        color="#D3D1C7",
        width=0.5,
        label="95% range",
        edgecolor="none"
    )

    ax2.bar(
        x,
        fc_ci.iloc[:, 1].values - fc_ci.iloc[:, 0].values,
        bottom=fc_ci.iloc[:, 0].values,
        color="#85B7EB",
        width=0.5,
        label="80% range",
        edgecolor="none"
    )

    ax2.plot(
        x,
        fc_mean.values,
        "o-",
        color="#0C447C",
        lw=2,
        label="Base case"
    )

    ax2.set_xticks(x)

    ax2.set_xticklabels(
        fc_x.strftime("%b %Y")
    )

    ax2.yaxis.set_major_formatter(
        mticker.FuncFormatter(lambda v, _: f"${v:,.0f}")
    )

    ax2.set_title(
        "Monthly Forecast: Floor / Base / Ceiling",
        fontweight="bold"
    )

    ax2.legend(fontsize=9)

    # =====================================================
    # FINAL FORMATTING
    # =====================================================
    fig.tight_layout()

    plt.show()

    # =====================================================
    # SUMMARY
    # =====================================================
    print(f"\nForecast summary (next {FORECAST_PERIODS} months):")

    print(f"  Base case total  : ${fc_mean.sum():,.0f}")
    print(f"  Floor (95% low)  : ${fc_ci_wide.iloc[:,0].sum():,.0f}")
    print(f"  Ceiling (95% hi) : ${fc_ci_wide.iloc[:,1].sum():,.0f}")
    print(f"  Model AIC        : {result.aic:.1f}")

except Exception as e:
    print(f"SARIMA failed: {e}")
    print("Need >= 24 months of clean monthly numeric data for reliable forecasting.")
```

## Inference

- The SARIMA model projects **H1 2015 retail revenue** at approximately:

  - **Base case:** ~$2.27M  
  - **Realistic floor:** ~$1.54M  
  - **Realistic ceiling:** ~$2.99M  

- This implies a relatively wide **uncertainty band of ~$1.45M**, highlighting meaningful forecast variability.

- The model captures a clear **upward seasonal trend** that has been visible since 2013.

- Among the forecasted months, **June 2015** is expected to be the strongest performer, with projected revenue of approximately:

  **~$533K**

- Given the breadth of the confidence interval, the business should:
  - plan inventory and working capital conservatively, closer to the **$1.5M floor**, while
  - treating the upper range as an **upside scenario dependent on strong peak-season execution**.

> **Operational planning should prioritize downside resilience while retaining flexibility to capture seasonal demand upside.**

## QUESTION 2

Which segment will drive the most growth next quarter — and which is at risk?

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle(
    "Q2 — Which segment will drive the most growth next quarter — and which is at risk?",
    fontsize=12, fontweight="bold"
)

seg_forecasts = {}
for seg, colour in SEG_PALETTE.items():
    s = seg_ts.get(seg)
    if s is None or len(s) < 18:
        continue
    try:
        m  = SARIMAX(s, order=(1,1,1), seasonal_order=(1,1,0,12),
                     enforce_stationarity=False, enforce_invertibility=False)
        r  = m.fit(disp=False)
        fc = r.get_forecast(steps=FORECAST_PERIODS)
        seg_forecasts[seg] = {
            "mean": fc.predicted_mean,
            "ci":   fc.conf_int(alpha=0.2),
        }
        # plot on left: actual + forecast per segment
        axes[0].plot(s.index.to_timestamp(), s.values,
                     color=colour, lw=1.2, alpha=0.5)
        axes[0].plot(fc.predicted_mean.index.to_timestamp(),
                     fc.predicted_mean.values,
                     color=colour, lw=2.2, ls="--", label=seg)
        axes[0].fill_between(
            fc.predicted_mean.index.to_timestamp(),
            fc.conf_int(alpha=0.2).iloc[:,0],
            fc.conf_int(alpha=0.2).iloc[:,1],
            alpha=0.1, color=colour
        )
    except Exception:
        pass

axes[0].axvline(ts.index[-1].to_timestamp(), color="#888", lw=1, ls=":")
axes[0].yaxis.set_major_formatter(mticker.FuncFormatter(lambda v, _: f"${v:,.0f}"))
axes[0].set_title("Segment revenue — actual + forecast", fontweight="bold")
axes[0].legend(fontsize=9)

# right: forecast period totals + implied growth vs last 6 months
labels, fwd_totals, hist_totals, colours_bar = [], [], [], []
for seg, colour in SEG_PALETTE.items():
    if seg not in seg_forecasts:
        continue
    s         = seg_ts[seg]
    hist_6    = s.iloc[-6:].sum()
    fwd_total = seg_forecasts[seg]["mean"].sum()
    labels.append(seg)
    hist_totals.append(hist_6)
    fwd_totals.append(fwd_total)
    colours_bar.append(colour)

x = np.arange(len(labels))
axes[1].bar(x - 0.2, hist_totals, width=0.35, color="#D3D1C7", edgecolor="none", label="Last 6 months")
axes[1].bar(x + 0.2, fwd_totals,  width=0.35, color=colours_bar, alpha=0.85, edgecolor="none", label="Next 6 months (forecast)")
axes[1].set_xticks(x); axes[1].set_xticklabels(labels)
axes[1].yaxis.set_major_formatter(mticker.FuncFormatter(lambda v, _: f"${v:,.0f}"))
axes[1].set_title("Revenue: last 6 months vs next 6 months forecast", fontweight="bold")
axes[1].legend(fontsize=9)

for i, (h, f) in enumerate(zip(hist_totals, fwd_totals)):
    pct = (f - h) / h * 100 if h else 0
    axes[1].text(i + 0.2, f + max(fwd_totals)*0.01,
                 f"{pct:+.0f}%", ha="center", fontsize=9, fontweight="bold",
                 color="#1D9E75" if pct >= 0 else "#D85A30")

fig.tight_layout()
plt.show()

print("\nSegment forecast vs prior period:")
for seg, h, f in zip(labels, hist_totals, fwd_totals):
    pct = (f - h) / h * 100 if h else 0
    flag = "growing" if pct >= 0 else "AT RISK"
    print(f"  {seg:15s}: last 6m ${h:,.0f}  ->  next 6m ${f:,.0f}  ({pct:+.0f}%)  [{flag}]")
```

## Inference

- All three customer segments are forecast to decline by approximately **10% next quarter**:

  - **Consumer:** ↓ ~$141K  
  - **Corporate:** ↓ ~$80K  
  - **Home Office:** ↓ ~$53K  

- The pattern suggests a **broad-based demand softening** rather than a segment-specific weakness.

- No segment currently appears positioned to drive near-term growth; all segments can be considered:

  **"At risk"**

- Among them, **Consumer** represents the most critical exposure due to its scale, with an expected forecast contribution of approximately:

  **~$1.2M in revenue**

- As the largest revenue contributor, Consumer also carries the highest absolute downside risk.

- Immediate strategic priority should focus on:
  - targeted retention initiatives,
  - customer reactivation campaigns, and
  - demand stimulation efforts within the Consumer segment ahead of the forecast period.

> **Stabilizing Consumer demand is likely to have the greatest impact on mitigating overall revenue decline.**

## QUESTION 3

How much of projected revenue is real growth vs seasonal lift we shouldn't bank on?

```python
if len(ts) >= 24:
    stl = STL(ts, period=12, robust=True).fit()

    fig, axes = plt.subplots(3, 1, figsize=(14, 10), sharex=True)
    fig.suptitle(
        "Q3 — How much of projected revenue is real growth vs seasonal lift we shouldn't bank on?",
        fontsize=12, fontweight="bold"
    )

    idx = ts.index.to_timestamp()

    axes[0].plot(idx, stl.trend, color="#378ADD", lw=2, label="Trend")
    axes[0].fill_between(idx, stl.trend, alpha=0.1, color="#378ADD")
    axes[0].yaxis.set_major_formatter(mticker.FuncFormatter(lambda v, _: f"${v:,.0f}"))
    axes[0].set_title("Structural trend — the 'true' underlying growth", fontweight="bold")
    axes[0].legend(fontsize=9)

    colours_s = ["#1D9E75" if v >= 0 else "#D85A30" for v in stl.seasonal]
    axes[1].bar(idx, stl.seasonal, color=colours_s, width=20, edgecolor="none")
    axes[1].axhline(0, color="#888", lw=0.8)
    axes[1].yaxis.set_major_formatter(mticker.FuncFormatter(lambda v, _: f"${v:,.0f}"))
    axes[1].set_title("Seasonal component — predictable calendar effect (green = boost, red = drag)", fontweight="bold")

    axes[2].bar(idx, stl.resid, color="#7F77DD", width=20, alpha=0.7, edgecolor="none")
    axes[2].axhline(0, color="#888", lw=0.8)
    axes[2].yaxis.set_major_formatter(mticker.FuncFormatter(lambda v, _: f"${v:,.0f}"))
    axes[2].set_title("Residual — unexplained noise (investigate large spikes)", fontweight="bold")

    fig.tight_layout()
    plt.show()

    seasonal_pct = np.abs(stl.seasonal).mean() / ts.mean() * 100
    trend_pct    = np.abs(stl.trend  ).mean() / ts.mean() * 100
    resid_pct    = np.abs(stl.resid  ).mean() / ts.mean() * 100
    peak_lift    = stl.seasonal.max()
    trough_drag  = stl.seasonal.min()

    print(f"\nVariance decomposition (avg absolute contribution):")
    print(f"  Trend    : {trend_pct:.1f}% of avg monthly revenue")
    print(f"  Seasonal : {seasonal_pct:.1f}%  (peak lift ${peak_lift:,.0f}  |  trough drag ${trough_drag:,.0f})")
    print(f"  Residual : {resid_pct:.1f}%  (unexplained)")
    print(f"\n  Interpretation: {seasonal_pct:.0f}% of a typical month's revenue")
    print(f"  is seasonal — do not confuse Q4 spike with structural growth.")
else:
    print("Need >= 24 months for STL decomposition.")
```

## Inference

- The business demonstrates clear and consistent **underlying structural growth**.

- Trend-adjusted monthly revenue has increased from approximately:

  - **~$170K/month in 2011**
  - to **~$390K/month currently**

- However, a significant portion of monthly revenue variation is driven by seasonality:
  
  - Approximately **29% of monthly revenue** reflects seasonal fluctuation rather than core growth.

- Seasonal effects are particularly pronounced across quarters:

  - **Q4 seasonal uplift:** ~+$167K  
  - **Q1 seasonal drag:** ~−$139K  

- As a result, the sharp revenue spikes observed in **late 2013** appear to be driven primarily by:
  - seasonal demand patterns,
  - rather than permanent structural acceleration.

- This has important implications for planning and performance management:
  - budgeting should rely on **trend-adjusted baselines**,
  - and forward targets should account for recurring seasonal effects.

> **Treating seasonal Q4 uplifts as permanent growth risks overstating business momentum and inflating future performance expectations.**

## QUESTION 4

Which leading indicators predict a revenue dip early enough to act on?

```python
lag_clean = lag_df.dropna()

corr_cols = [
    "lag_1",
    "lag_2",
    "lag_3",
    "rolling_3m_avg",
    "mom_growth",
    "qoq_growth",
    "is_q4"
]

corr_with_revenue = (
    lag_clean[corr_cols]
    .corrwith(lag_clean["revenue"])
    .sort_values()
)

# =========================================================
# CREATE FIGURE
# =========================================================
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

fig.suptitle(
    "Q4 — Which leading indicators predict a revenue dip early enough to act on?",
    fontsize=12,
    fontweight="bold"
)

# =========================================================
# LEFT PLOT — Correlation strength
# =========================================================
colours_c = [
    "#D85A30" if v < 0 else "#1D9E75"
    for v in corr_with_revenue.values
]

axes[0].barh(
    corr_with_revenue.index,
    corr_with_revenue.values,
    color=colours_c,
    edgecolor="none",
    height=0.55
)

axes[0].axvline(0, color="#888", lw=0.8)

axes[0].axvline(
    0.7,
    color="#1D9E75",
    lw=1,
    ls="--",
    alpha=0.6
)

axes[0].axvline(
    -0.7,
    color="#D85A30",
    lw=1,
    ls="--",
    alpha=0.6
)

axes[0].set_xlabel(
    "Pearson correlation with next-month revenue"
)

axes[0].set_title(
    "Leading indicator strength",
    fontweight="bold"
)

axes[0].set_xlim(-1, 1)

# Correlation labels
for i, (label, val) in enumerate(corr_with_revenue.items()):

    axes[0].text(
        val + (0.03 if val >= 0 else -0.03),
        i,
        f"{val:.2f}",
        va="center",
        ha="left" if val >= 0 else "right",
        fontsize=9
    )

# =========================================================
# RIGHT PLOT — Strongest predictor vs revenue
# =========================================================
best_pred = corr_with_revenue.abs().idxmax()

axes[1].scatter(
    lag_clean[best_pred],
    lag_clean["revenue"],
    color="#378ADD",
    alpha=0.5,
    s=25,
    edgecolors="none"
)

# Regression line
coef = np.polyfit(
    lag_clean[best_pred].values,
    lag_clean["revenue"].values,
    1
)

x_fit = np.linspace(
    lag_clean[best_pred].min(),
    lag_clean[best_pred].max(),
    100
)

axes[1].plot(
    x_fit,
    np.poly1d(coef)(x_fit),
    color="#D85A30",
    lw=2
)

# =========================================================
# IMPROVED AXIS FORMATTING
# =========================================================

# X-axis formatting
axes[1].xaxis.set_major_formatter(
    mticker.FuncFormatter(lambda v, _: f"${v/1000:.0f}K")
)

axes[1].xaxis.set_major_locator(
    mticker.MaxNLocator(6)
)

axes[1].tick_params(
    axis="x",
    rotation=15
)

axes[1].grid(
    axis="x",
    alpha=0.15
)

# Y-axis formatting
axes[1].yaxis.set_major_formatter(
    mticker.FuncFormatter(lambda v, _: f"${v/1000:.0f}K")
)

# Labels & title
axes[1].set_xlabel(
    f"{best_pred} (predictor)"
)

axes[1].set_ylabel(
    "Actual revenue"
)

axes[1].set_title(
    f"Strongest predictor: {best_pred} (r={corr_with_revenue[best_pred]:.2f})",
    fontweight="bold"
)

# =========================================================
# FINAL LAYOUT
# =========================================================
fig.tight_layout()

plt.show()

# =========================================================
# PRINT RANKING
# =========================================================
print("\nLeading indicator ranking (by absolute correlation with revenue):")

ranked = corr_with_revenue.abs().sort_values(ascending=False)

label_map = {
    "lag_1":          "Revenue 1 month ago",
    "lag_2":          "Revenue 2 months ago",
    "lag_3":          "Revenue 3 months ago",
    "rolling_3m_avg": "3-month rolling avg (lagged)",
    "mom_growth":     "Month-on-month growth rate",
    "qoq_growth":     "Quarter-on-quarter growth rate",
    "is_q4":          "Is Q4 flag (Oct-Dec)"
}

for col, val in ranked.items():

    strength = (
        "strong" if val >= 0.7
        else "moderate" if val >= 0.4
        else "weak"
    )

    print(
        f"  {label_map.get(col, col):38s}: "
        f"r={corr_with_revenue[col]:+.2f}  [{strength}]"
    )

print(
    f"\nMonitor '{best_pred}' as your primary early warning KPI."
)
```

## Inference

- Revenue momentum appears highly persistent over multi-month periods.

- When revenue remains strong for **2–3 consecutive months**, the following month is also likely to perform strongly — and the reverse pattern holds during downturns.

- The strongest early warning indicator identified is the:

  **3-month rolling average revenue (r = 0.69)**

- This predictor outperforms:
  - last month's revenue alone,
  - month-on-month growth rates, and
  - quarter-on-quarter growth trends.

- In contrast, short-term growth metrics demonstrate relatively weak predictive reliability for future revenue performance.

- From an operational perspective, the business should closely monitor rolling revenue momentum rather than isolated monthly fluctuations.

- A practical intervention framework could be:

  - If the **3-month rolling average declines for two consecutive months**:
    - trigger promotional campaigns,
    - increase customer outreach,
    - adjust inventory planning, and
    - activate demand-generation efforts proactively.

> **Sustained declines in rolling revenue should be treated as an early-stage business risk signal rather than waiting for a full revenue shortfall to emerge.**
