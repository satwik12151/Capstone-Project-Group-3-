```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/satwik12151/Capstone-Project-Group-3-/main/Global_Superstore2.csv"

df = pd.read_csv(url, encoding='latin1')

df.head()
```

```python
df.describe(include='all')
```

```python
df.info()
```

```python
#Better processing and grouping of data
cat_cols = [
    'Ship Mode', 'Segment', 'Market',
    'Region', 'Category', 'Sub-Category',
    'Order Priority'
]

for col in cat_cols:
    df[col] = df[col].astype('category')
```

```python
## Particular row is waste as we have order unique row id
df.drop('Row ID', axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=['Postal Code'])
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
text_columns = [
    'City',
    'State',
    'Country',
    'Region',
    'Segment',
    'Category',
    'Sub-Category',
    'Product Name'
]

for col in text_columns:
    df[col] = df[col].str.strip().str.title()
```

```python
df[df['Row ID'].duplicated()]
```

```python
#Cheking for invalid values
(df['Quantity'] <= 0).sum()
(df['Sales'] <= 0).sum()
```

```python
#Cheking for Categorical label consistency
print(df['Category'].unique())
print(df['Region'].unique())
print(df['Ship Mode'].unique())
```

```python
print(df['Category'].value_counts())
print(df['Region'].value_counts())
print(df['Ship Mode'].value_counts())
```

####Shipping dates less than order date is data corruption looking for reasons and correcting that

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)

df['Ship Date'] = pd.to_datetime(df['Ship Date'], dayfirst=True)
```
